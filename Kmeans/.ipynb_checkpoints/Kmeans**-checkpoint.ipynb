```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(16,9)
plt.style.use('ggplot')
import warnings
```

```python
data=pd.read_csv('./Datasets/xclara.csv')
```

```python
print(data.shape)
```

```python
data.head(n=5)
```

```python
f1=data['V1'].values
f2=data['V2'].values
```

```python
X=np.array(list(zip(f1,f2)))
plt.scatter(X[:,0],X[:,1],c='blue',s=10)
```

```python
k=8
color=["green","red","blue","yellow","orange","grey","maroon","salmon"]
clusters={}
for i in range(k):
    center=10*(2*np.random.random((X.shape[1],))-1)
    points=[]
    cluster={
        'center':center,
        'points':points,
        'color' : color[i]
    }
    clusters[i]=cluster
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def assignPointToClusters(clusters):
    for ix in range(X.shape[0]):
        dist=[]
        curr_x=X[ix]
        for kx in range(k):
            d=distance(curr_x,clusters[kx]['center'])
            dist.append(d)
        current_cluster=np.argmin(dist)
        clusters[current_cluster]['points'].append(curr_x)
```

```python
def updateCluster(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        if pts.shape[0]>0:
            new_u=pts.mean(axis=0)
            clusters[kx]['center']=new_u
            clusters[kx]['points']=[]
```

```python
def plotClusters(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'],s=10)
        except:
            pass
        uk=clusters[kx]['center']
        plt.scatter(uk[0],uk[1],color='k',marker="*")
```

```python
assignPointToClusters(clusters)
plotClusters(clusters)
```

```python
assignPointToClusters(clusters)
plotClusters(clusters)
updateCluster(clusters)
```
