- kmeans algorithm is an unsupervised learning based algorithm it basically club similar type of samples

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

- Data extract from sklearn library

```python
X,Y=make_blobs(n_features=2,n_samples=1000,centers=8,random_state=3)
```

### Data Visualization

```python
plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

- Here I am making dictonary of clusters which contains center,points of samples and color of each cluster 

```python

k=8
color=["green","red","blue","yellow","orange","grey","maroon","salmon"]
clusters={}
for i in range(k):
    center=10*(2*np.random.random((X.shape[1],))-1)
    points=[]
    cluster={
        'center':center,
        'points':points,
        'color' : color[i]
    }
    clusters[i]=cluster
```

- Euclidean Distance

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

- Assigning each samples points according to their cluster centers
- This is called as E step 

```python
def assignPointToClusters(clusters):
    for ix in range(X.shape[0]):
        dist=[]
        curr_x=X[ix]
        for kx in range(k):
            d=distance(curr_x,clusters[kx]['center'])
            dist.append(d)
        current_cluster=np.argmin(dist)
        clusters[current_cluster]['points'].append(curr_x)
```

- Update clusters according to the means of points
- This is called as M step

```python
def updateCluster(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        if pts.shape[0]>0:
            new_u=pts.mean(axis=0)
            clusters[kx]['center']=new_u
            clusters[kx]['points']=[]
```

- Plot Function

```python
def plotClusters(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
        except:
            pass
        uk=clusters[kx]['center']
        plt.scatter(uk[0],uk[1],color='k',marker="*")
```

- This is before applying kmeans

```python
assignPointToClusters(clusters)
plotClusters(clusters)
```

- This is after applying kmeans

```python
assignPointToClusters(clusters)
plotClusters(clusters)
updateCluster(clusters)
```
