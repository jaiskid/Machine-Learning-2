# K means 3 D 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.rcParams['figure.figsize'] = (16, 9)
```

```python
X, y = make_blobs(n_samples=1500, n_features=3, centers=5)
```

```python
k=5
color=["red","yellow","orange","grey","salmon"]
clusters={}
for i in range(k):
    center=10*(2*np.random.random((X.shape[1],))-1)
    points=[]
    cluster={
    'center':center,
    'points':points,
    'color': color[i]
    }
    clusters[i]=cluster
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def assignPointsClusters(clusters):
    for ix in range(X.shape[0]):
        dist=[]
        curr_x=X[ix]
        for kx in range(k):
            d=distance(curr_x,clusters[kx]['center'])
            dist.append(d)
            current_cluster=np.argmin(dist)
            clusters[current_cluster]['points'].append(curr_x)
```

```python
def updateCluster(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        if pts.shape[0]>0:
            new_u=pts.mean(axis=0)
            clusters[kx]['center']=new_u
            clusters[kx]['points']=[]
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2],s=5)
```

```python
def plotClusters(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        try:
            ax.scatter(pts[:,0],pts[:,1],pts[:,2],color=clusters[kx]['color'],s=5)
        except:
            pass
        uk=clusters[kx]['center']
        print(uk)
        ax.scatter(uk[0],uk[1],uk[2],marker='*', c='#050505', s=100)
    plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
assignPointsClusters(clusters)
plotClusters(clusters)
```

```python
for i in range(8):
    fig = plt.figure()
    ax = Axes3D(fig)
    assignPointsClusters(clusters)
    plotClusters(clusters)
    updateCluster(clusters)
```
